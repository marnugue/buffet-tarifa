```python
import blurhash
import numpy as np
import PIL.Image
img = np.array(PIL.Image.open("../public/DSC_0001.JPG").convert("RGB"))
UBL = blurhash.encode(img)
# Result: 'UBL_:rOpGG-oBUNG,qRj2so|=eE1w^n4S5NH'
```

```python
PIL.Image.fromarray(np.array(blurhash.decode(UBL,img.shape[0], img.shape[1])).astype('uint8'))
```
